```python
# Creating categories for rankings
rating_score = []

for score in df_new['rating']:
    if score < 2.0 : rating_score.append('Poor')         # Score below 2.0
    elif score < 3.0 : rating_score.append('Fair')       # Score range of 2.0 - 2.9
    elif score < 4.0 : rating_score.append('Average')    # Score range of 3.0 - 3.9
    elif score < 5.0 : rating_score.append('Good')       # Score range of 4.0 - 4.9
    elif score == 5.0 : rating_score.append('Excellent') # Score of 5.0
```

```python
# Creating a new column 
df_new['rating_score'] = rating_score 

# Changing the data type
df_new['rating_score'] = df_new['rating_score'].astype('category')
```

```python
#

# Creating figure and axes
fig, ax = plt.subplots(figsize = (20, 5))
ax = sns.countplot(x = df_new['rating_score'], palette = 'mako', order = ['Poor', 'Fair', 'Average', 'Good', 'Excellent'])

# Adding titles to axes and graph
ax.set_xlabel('Rating score', fontweight = 'bold')
ax.set_ylabel('Counts', fontweight = 'bold')
ax.set_title('RATING SCORE IN NUMBERS', fontweight = 'heavy', size = 'x-large')

# Adding labels to bins
ax.bar_label(container = ax.containers[0])

# Visualization
plt.show()

# Most of the products in this dataset have been rated 'Good'. 
# There are extremely few products that are rated 'Fair' and 'Excellent'. 
# No product has a 'Poor' rating in this dataset.
```
